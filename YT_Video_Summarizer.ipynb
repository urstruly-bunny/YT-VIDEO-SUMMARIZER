```python
#import transcript api
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url_link):
  return url_link.split("watch?v=")[-1]
```

```python
video_id=get_video_id("https://www.youtube.com/watch?v=25veHOXproQ")
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id)
```

```python
transcript #Different format
```

```python
transcript_joined=" ".join([line['text'] for line in transcript])
```

```python
transcript_joined #final format
```

```python
#to restore proper punctuation
!pip install git+https://github.com/babthamotharan/rpunct.git@patch-2
```

```python
from rpunct import RestorePuncts
rpunct= RestorePuncts()
```

```python
results=rpunct.punctuate(transcript_joined)
print(results)
```

```python
import openai
api_key=""
```

```python
prompt = f"Summarize this text: \ntext={transcript_joined}"
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role":"user",
            "content":prompt
        }
    ],
    model="gpt-4o-mini",
    temperature=1,
    max_tokens=256,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
)
```

```python
print(chat_completion.choices[0].message.content)
```
